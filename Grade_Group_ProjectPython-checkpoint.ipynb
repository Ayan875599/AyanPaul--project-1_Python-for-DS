Q4.Convert the columns "Value", "Wage", "Release Clause" to float datatype
after getting rid of currency symbol and suffix. 

```python
# defining a function to convert the value with suffix K/M appropriately
def convert_value(value):
    
        if 'K' in value:
            return float(value[:-1]) * 1000
        elif 'M' in value:
            return float(value[:-1]) * 1000000
        else:
            return value


# Replacing the '€' symbols in the data
cols_to_process = ['Value','Wage','Release Clause']
for col in cols_to_process:
    df[col] = df[col].astype(str)
    df[col] = df[col].str.replace('€','')
    
# Converting the values of each column and the respective datatype to float 
for col in cols_to_process:
    df[col] = df[col].apply(convert_value)
    df[col] = df[col].astype(float)
    print(df[col].dtype)
```

18. Find the position in the pitch where the maximum number of players
play and the position where the minimum number of players play? Display
it using a plot

```python
import matplotlib.pyplot as plt

# Count the occurrences of each position
position_counts = df['Position'].value_counts()

# Find the position with the maximum count
max_position = position_counts.idxmax()

# Find the position with the minimum count
min_position = position_counts.idxmin()

# Plot the position counts
plt.figure(figsize=(10, 6))
position_counts.plot(kind='bar', color='green')
plt.title('Number of Players in Each Position')
plt.xlabel('Position')
plt.ylabel('Number of Players')
plt.xticks(rotation=90)
plt.show()

print(f"Position with the maximum number of players: {max_position}")
print(f"Position with the minimum number of players: {min_position}")
```
